```python
#the beer song
def bottles(x):
    for num in range(x, 0, -1):
        if num == 1:
            print("""{x} bottle of beer on the wall, {x} bottle of beer.
            Take one down, pass it around,""".format(x = num))
        while num > 1:
            print("""{x} bottles of beer on the wall, {x} bottles of beer. 
            Take one down, pass it around,""".format(x = num))
            break

    print("""No more bottles of beer on the wall, no more bottles of beer. 
            Go to the store and buy some more {x} bottles of beer on the wall...""".format(x = x))

bottles(99)
```
